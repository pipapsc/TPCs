## GrÃ¡ficos em forma de tarte

```python
import matplotlib.pyplot as plt
 
# defining labels
activities = ['eat', 'sleep', 'work', 'play']
 
# portion covered by each label
slices = [3, 7, 8, 6]
 
# color for each label
colors = ['r', 'y', 'g', 'b']
 
# plotting the pie chart
plt.pie(slices, labels = activities, colors=colors,
        startangle=90, shadow = True, explode = (0, 0, 0.1, 0),
        radius = 1.2, autopct = '%1.1f%%')
 
# plotting legend
plt.legend()
 
# showing the plot
plt.show()
```

```python
print("""Menu:
        (1) Criar Polinomio
        (2) Calcular Polinomio
        (3) Calcular Tabela
        (4) Simplificar Polinomio
        (5) Derivar Polinomio
        (6) Representar o polinómio graficamente
        (0) Sair""")

def sair():
    print("That's all folks :)))")

def criarTermo():
    coeficiente = int(input("Introduza o coeficiente: "))
    expoente = int(input("Introduza o expoente: "))
    t = (coeficiente, expoente)
    return t

def inserirTermo(p, t):
    return p.append(t)

def criarPolinomio():
    n = int(input("Insira o comprimento do polinomio: "))
    pol = []
    i = 0
    while i < n:
        pol.append(criarTermo())
        i += 1
    return pol

def verTermo(t):
    c, e = t
    return str(c)+'x^'+str(e)

def verPolinomio(p):
    res = ""
    for t in p:
        if res == "":
            res = verTermo(t)
        else:
            res = res + " + " + verTermo(t)
    print(res)
    
def calcularTermo(t, x):  
    c, e = t
    res = c*(x**e)
    return res

def calculaPolinomio(p, x):
    resFinal = 0
    for t in p:
        polRes = calcularTermo(t, x)
        resFinal += polRes
    print(resFinal)
    
def calculaTabela(p, nLinhas):
    for i in range(nLinhas + 1):
        print(str(i) + " .... " + str(calculaPolinomio(p, i)))

def chaveOrd(t):
    _,e = t # _e quando o elemento nao interessa
    return e
        
def ordenarPolinomio(p):
    p.sort(reverse = True, key = chaveOrd) 
        
def simplificarPolinomio(p):
    ordenarPolinomio(p)
    resLista = []
    for i in range(1, len(p)):
        c, e = p[i]
        c0, e0 = p[i-1]
        if e0 == e:
            inserirTermo(resLista, (c+c0, e))
        else: 
            if e0 > e:
                if i == 1:
                    inserirTermo(resLista, (c0, e0))
            else:
                inserirTermo(resLista, (c,e))
    return resLista

def derivarPolinomio(p):
    res = []
    for (c,e) in p:
        if e > 1:
            res.append((c*e, e-1))
    return res

def graficoPol(p):
    x = [i for i in range(-100,100)]
    y = []
    for n in x:
        y.append(calculaPolinomio(p, n))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('p(x)')
    plt.title('Polinomio')
        
def opcao():
    op = str(input("Insira um dos comandos do menu: "))
    while op != '0':
        if op == '1':
            polinomio = criarPolinomio()
            verPolinomio(polinomio)
        elif op == '2':
            x = int(input("Insira o valor de X: "))
            calculaPolinomio(polinomio, x)
        elif op == '3':
            numeroLinhas = int(input("Insira o numero de Linhas: "))
            calculaTabela(polinomio, numeroLinhas)
        elif op == '4':
            polinomioSimplificado = simplificarPolinomio(polinomio)
            verPolinomio(polinomioSimplificado)
        elif op == '5':
            novoPolinomio = derivarPolinomio(polinomio)
            verPolinomio(novoPolinomio)
        elif op == '6':
            graficoPol(polinomio)
        else:
            print("Este numero nao esta nas alternativas, tente novamente :D.\n")
        op = str(input("Insira um dos comandos do menu: "))
    if op == '0':
        sair()
opcao()

```
