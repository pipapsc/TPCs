### Plot1: uma espÃ©cie de "Hello world!"

```python
# importing the required module
import matplotlib.pyplot as plt
 
# x axis values
x = [1,2,3]
# corresponding y axis values
y = [2,4,1]
 
# plotting the points
plt.plot(x, y)
 
# naming the x axis
plt.xlabel('Abcissas')
# naming the y axis
plt.ylabel('Ordenadas')
 
# giving a title to my graph
plt.title('O meu primeiro grÃ¡fico!')
 
# function to show the plot
plt.show()
```

### Plot2: duas funÃ§Ãµes no mesmo grÃ¡fico

```python
import matplotlib.pyplot as plt
 
# line 1 points
x1 = [1,2,3]
y1 = [2,4,1]
# plotting the line 1 points
plt.plot(x1, y1, label = "linha 1")
 
# line 2 points
x2 = [1,2,3]
y2 = [4,1,3]
# plotting the line 2 points
plt.plot(x2, y2, label = "linha 2")
 
# naming the x axis
plt.xlabel('Abcissas')
# naming the y axis
plt.ylabel('Ordenadas')
# giving a title to my graph
plt.title('GrÃ¡fico com duas funÃ§Ãµes')
 
# show a legend on the plot
plt.legend()
 
# function to show the plot
plt.show()
```

### Plot3: configuraÃ§Ãµes

```python
import matplotlib.pyplot as plt
 
# x axis values
x = [1,2,3,4,5,6]
# corresponding y axis values
y = [2,4,1,5,2,6]
 
# plotting the points
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=12)
 
# setting x and y axis range
plt.ylim(1,8)
plt.xlim(1,8)
 
# naming the x axis
plt.xlabel('Abcissas')
# naming the y axis
plt.ylabel('Ordenadas')
 
# giving a title to my graph
plt.title('Um grÃ¡fico com estilo!')
 
# function to show the plot
plt.show()
```

## GrÃ¡ficos em forma de tarte

```python
import matplotlib.pyplot as plt
 
# defining labels
activities = ['eat', 'sleep', 'work', 'play']
 
# portion covered by each label
slices = [3, 7, 8, 6]
 
# color for each label
colors = ['r', 'y', 'g', 'b']
 
# plotting the pie chart
plt.pie(slices, labels = activities, colors=colors,
        startangle=90, shadow = True, explode = (0, 0, 0.1, 0),
        radius = 1.2, autopct = '%1.1f%%')
 
# plotting legend
plt.legend()
 
# showing the plot
plt.show()
```
